How did country ranks or scores change between the 2015 and 2016 reports?

```python
h1 = df[['Country','Happiness Rank','Happiness Score']]
#h2 = df2.loc[:,'Country':'Happiness Score']
#h2 = df2[df2.columns[0:7]] 
h2 = df2[['Country','Happiness Rank','Happiness Score']]
print(h1.head(5))
print(h2.head(5))
```

```python
join = pd.merge(h1,h2, on='Country', how ='left')
join.head(5)
```

```python
Hlist = join.rename(columns={'Happiness Rank_x': 'HR2015', 'Happiness Score_x': 'HScore2015', 'Happiness Rank_y':'HR2016', 'Happiness Score_y':'HScore2016'})
Hlist.head(5)
```

```python
Hlist['Rank Diff'] = Hlist['HR2015'] - Hlist['HR2016'] #the best is low number
Hlist['Score Diff'] = Hlist['HScore2015'] - Hlist['HScore2016']  #the latest minus the older data
Hlist.head()
```

Did any country experience a significant increase or decrease in happiness?

```python
Hlist.sort_values(['Rank Diff','Score Diff'], ascending = False)
```

```python
Hlist.sort_values(['Rank Diff','Score Diff'], ascending = True)
```
