```python
h1 = happy2015[['Country','Happiness Rank','Happiness Score']]
h1.head()
```

```python
htest = happy2015.loc[:,'Country':'Happiness Score']
htest.head()
```

```python
h2 = happy2016[['Country','Happiness Rank','Happiness Score']]
h2.head()
```

Join table h1 and h2

```python
Hlist = pd.merge(h2, h1, on='Country', how='left')
Hlist.head()       
```

Change column name

```python
Hlist = Hlist.rename(columns={'Happiness Rank_x': 'HR2016', 'Happiness Score_x': 'HScore2016', 'Happiness Rank_y':'HR2015', 'Happiness Score_y':'HScore2015'})
```

```python
Hlist.head()
```

Create a new column called Rank Diff & Score Diff

```python
Hlist['Rank Diff'] = Hlist['HR2015'] - Hlist['HR2016'] #the best is low number
Hlist['Score Diff'] = Hlist['HScore2016'] - Hlist['HScore2015']  #the latest minus the older data
Hlist.head()
```

Show the most improvement country by ranks only

```python
Hlist.sort_values('Rank Diff', ascending=False)
```

Show the most improvement country by ranks and scores

```python
Hlist.sort_values(['Rank Diff','Score Diff'], ascending=False)
```

Show the most backward country by ranks

```python
Hlist.sort_values('Rank Diff', ascending=True)
```

Show the most backward improvement by scores

```python
Hlist.sort_values('Score Diff', ascending=True)
```

Show the most improvement by scores

```python
Hlist.sort_values('Score Diff', ascending=False)
```
